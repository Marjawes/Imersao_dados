```python
cod_compostos = dados['composto'].value_counts(normalize=True).index[0:5]
cod_compostos
```

```python
dados.query('composto in @cod_compostos')
```

```python
sns.set()
plt.figure(figsize=(10,6))
ax = sns.countplot(x = 'composto',
                   data=dados.query('composto in @cod_compostos'), 
                   order=dados.query('composto in @cod_compostos')['composto'].value_counts().index,
                   palette='rocket')
ax.set_title('Top5 Compostos mais usados')
ax.set_ylabel('Frequência')
ax.set_xlabel('Nome do Composto')
plt.show()
```
